```python
selected_columns = ['Weekly_Study_Hours', 'Attendance', 'Reading', 'Notes', 'Listening_in_Class', 'Grade', 'Sports_activity','Additional_Work']

stu_narrowed = student[selected_columns]
stu_narrowed
```

#Milestone #3 Update: Based on feedback, I am moving forward with converting the double letter grades into broader categories. This will helpt to create a more readible figure.

#removing rows with 6 as the 'notes' and 'Listening_in_Class'

#removing the row with 3 as the value for 'Attendance'


```python
new_grade = {
    'AA': 'Excellent',
    'BA': 'Good',
    'BB': 'Good',
    'CB': 'Average',
    'CC': 'Average',
    'DC': 'Below Average',
    'DD': 'Below Average',
    'Fail': 'Fail'
}


stu_clean = stu_narrowed[stu_narrowed['Notes'] != '6']
stu_clean = stu_clean[stu_clean['Listening_in_Class'] != '6']
stu_clean = stu_clean[stu_clean['Attendance'] != '3']

stu_clean['Grade'] = stu_clean['Grade'].map(new_grade)

print(stu_clean)
```

*When it comes to creating a visualization for this project, I want to make a visualization that is easily readable by my audiences, which are high school students and as well as other staff in my department. Since my data is mostly categorical, I will use categorical visualizations.

```python
grade_order = ['Excellent', 'Good', 'Average', 'Below Average', 'Fail']
fig1 = sns.countplot(data=stu_clean, x="Grade", hue="Attendance", order=grade_order)

# Set the labels
plt.xlabel('Grade')
plt.ylabel('Number of Students')

# Show the plot
plt.show()
```
